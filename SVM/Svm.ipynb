```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  let create the synthetic data point
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2,random_state=42, n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
plt.scatter(pd.DataFrame(X)[0], pd.DataFrame(X)[1], c=y, cmap='coolwarm')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
# Ensure that the cell defining X_train and y_train is executed before this cell
svc.fit(X_train, y_train)
```

```python
# predict the test data
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score ,classification_report, confusion_matrix
```

```python

print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')
```

```python
rbf.fit(X_train, y_train)
```

```python
y_pred = rbf.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
