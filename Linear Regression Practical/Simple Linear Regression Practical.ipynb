import some libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
```

```python
df = pd.read_csv('height-weight.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Height'], df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
```

```python
# finding collrelation between height and weight
df.corr()
```

```python
# serabian for visualization
import seaborn as sns
sns.pairplot(df)
 
```

```python
# independent and dependent features
x = df[['Weight']] # independent feature should be in dataframe or 2 dimensal array
y = df[["Height"]]
```

```python
# first step
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)    
# second step
```

```python
## standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_Train =scaler.fit_transform(x_train)
```

```python
X_test = scaler.transform(x_test)
X_test
```

```python
# apply linear regression to X_train and X_test
from sklearn.linear_model import LinearRegression
regression  = LinearRegression(n_jobs=-1)
regression.fit(X_Train, y_train)
```

```python
print("Coefficient: ", regression.coef_)
print("Intercept: ", regression.intercept_)
```

```python
# plot regression line with regression line
plt.scatter(X_Train, y_train, color='red')
plt.plot(X_Train,regression.predict(X_Train), color='blue')
```


predict for test data


```python
# prediction for test data 
y_pred = regression.predict(X_test)
```

```python
# perforamnce matrix
from sklearn.metrics import mean_squared_error,mean_absolute_error
print("Mean Squared Error: ", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
# R squared score
from sklearn.metrics import r2_score
print("R2 Score: ", r2_score(y_test, y_pred))
```
